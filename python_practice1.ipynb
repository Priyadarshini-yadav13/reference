###try except finally




```python
try:
    # Attempt to convert a string to an integer
    num = int("hello")
except ValueError:
    # Handle the case where the conversion fails
    print("Could not convert to an integer.")
finally:
    # This will always execute
    print("Finished attempting conversion.")

print("This line runs after the try-except-finally block.")
```

# Task

## Identify duplicated employee names

### Subtask:
Find the employee names that appear more than once in the dataset.


**Reasoning**:
Identify and count the number of duplicated employee names in the dataset.



```python
duplicated_names_count = df['EmployeeName'].duplicated().sum()
print(f"Number of duplicated employee names: {duplicated_names_count}")
```

**Reasoning**:
The previous command failed because the dataframe `df` was not defined. I will reload the dataframe and then proceed to find and count the duplicated employee names.



```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Muralimekala/python/master/Salaries.csv')
duplicated_names_count = df['EmployeeName'].duplicated().sum()
print(f"Number of duplicated employee names: {duplicated_names_count}")
```

## Group by employee name

### Subtask:
Group the DataFrame by the 'EmployeeName' column.


**Reasoning**:
Group the DataFrame by the 'EmployeeName' column as requested by the subtask.



```python
grouped_by_name = df.groupby('EmployeeName')
```

```python
print(grouped_by_name)
```

## Calculate the average of numeric columns

### Subtask:
For each group (employee name), calculate the average of all numeric columns except 'Id'.


**Reasoning**:
Identify numeric columns excluding 'Id' and calculate the mean of these columns for each employee name group.



```python
#to add all the numerical column and drop id column with if condition
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
if 'Id' in numeric_cols:
    numeric_cols.remove('Id')

average_numeric_values = grouped_by_name[numeric_cols].mean()
display(average_numeric_values.head())
```

## Create a new dataframe with aggregated data

### Subtask:
Construct a new DataFrame containing the employee names and their corresponding average numeric values.


**Reasoning**:
Reset the index of the average_numeric_values DataFrame to make 'EmployeeName' a column and store it in a new DataFrame.



```python
df_aggregated = average_numeric_values.reset_index()
display(df_aggregated.head())
```

## Create a new id

### Subtask:
Add a new column to the aggregated DataFrame to serve as a new identifier. This could be a simple sequential index.


**Reasoning**:
Reset the index of the aggregated DataFrame and assign it to a new 'New_ID' column.



```python
#to add a new column with existing column
df_aggregated = df_aggregated.reset_index(drop=True)
df_aggregated['New_ID'] = df_aggregated.index
display(df_aggregated.head())
```

## Display the resulting dataframe

### Subtask:
Show the first few rows of the new DataFrame with the aggregated data and the new ID.


**Reasoning**:
Display the first few rows of the aggregated DataFrame to show the result of the previous steps, including the new ID.



```python
display(df_aggregated.head())
```
